```python
# Función para eliminar outliers por IQR
def eliminar_outliers(dfx, columna):
    dfx[columna] = pd.to_numeric(dfx[columna], errors='coerce')
    dfx = dfx.dropna(subset=[columna])
    Q1 = dfx[columna].quantile(0.25)
    Q3 = dfx[columna].quantile(0.75)
    IQR = Q3 - Q1
    limite_inferior = Q1 - 1.5 * IQR
    limite_superior = Q3 + 1.5 * IQR
    return dfx[(dfx[columna] >= limite_inferior) & (dfx[columna] <= limite_superior)]

try:
    logging.info("Eliminando outliers")
    # Aplicar sobre columnas relevantes
    df_trus = eliminar_outliers(df_trus, "trip_distance")
    df_trus = eliminar_outliers(df_trus, "fare_amount")
except Exception as e:
    logging.error("Ocurrió un error durante la ejecución de la consulta", exc_info=True)
```

**Refined Layer (Datos agregados y KPIs)**

**Patrón de Demanda y Tiempos Pico**

```python
# Defnicon de franjas horarias
def clasificar_franja_horaria(hora):
    if 6 <= hora < 10:
        return 'Mañana (Pico)'
    elif 10 <= hora < 16:
        return 'Mediodía (Valle)'
    elif 16 <= hora < 20:
        return 'Tarde (Pico)'
    elif 20 <= hora < 24:
        return 'Noche (Valle)'
    else:
        return 'Madrugada'
```

```python
try:
    load_start = time.time()
    logging.info("Creando columnas derivadas...")
    df_ref = df_trips
    df_ref['hour'] = df_ref['pickup_datetime'].dt.hour
    df_ref['day_of_week'] = df_ref['pickup_datetime'].dt.day_name()
    df_ref['trip_duration'] = (df_ref['dropoff_datetime'] - df_ref['pickup_datetime']).dt.total_seconds() / 60

    logging.info("Clasificacion por franja horara...")
    df_ref['franja_horaria'] = df_ref['hour'].apply(clasificar_franja_horaria)

    logging.info("Resumen estadisticas KPIs por día y franja horaria")
    resumen = df_ref.groupby(['day_of_week', 'franja_horaria']).agg(
        cantidad_viajes=('pickup_datetime', 'count'),
        duracion_promedio_min=('trip_duration', 'mean'),
        tarifa_promedio_usd=('fare_amount', 'mean')
    ).reset_index()
except Exception as e:
    logging.error("Ocurrió un error durante la ejecución de la consulta", exc_info=True)
    
df_ref.head(5)
```

**Eficiencia Geográfica y Económica**

```python
# Analizar la eficiencia operativa y económica por zona o borough.
try:
    logging.info("Crear columna de duración en minutos y horas...")
    df_ref['trip_duration_min'] = (df_ref['dropoff_datetime'] - df_ref['pickup_datetime']).dt.total_seconds() / 60
    df_ref['trip_duration_hr'] = df_ref['trip_duration_min'] / 60

    logging.info("Filtrar datos inválidos...")
    df_valid = df_ref[
        (df_ref['trip_distance'] > 0) &
        (df_ref['fare_amount'] > 0) &
        (df_ref['trip_duration_min'] > 1)  # evitar viajes extremadamente cortos o erróneos
    ]

    logging.info("Crear columnas de KPIs...")
    df_valid['fare_per_mile'] = df_valid['fare_amount'] / df_valid['trip_distance']
    df_valid['fare_per_min'] = df_valid['fare_amount'] / df_valid['trip_duration_min']
    df_valid['speed_mph'] = df_valid['trip_distance'] / df_valid['trip_duration_hr']

    logging.info("Agrupando por zona y calcular promedios...")
    kpis_por_zona = df_valid.groupby(['Zone', 'Borough']).agg(
        cantidad_viajes=('fare_amount', 'count'),
        ingreso_promedio_milla=('fare_per_mile', 'mean'),
        ingreso_promedio_minuto=('fare_per_min', 'mean'),
        velocidad_promedio_mph=('speed_mph', 'mean')
    ).reset_index()

    logging.info("Ranking de zonas más rentables...")
    kpis_por_zona['ranking_rentabilidad'] = kpis_por_zona['ingreso_promedio_milla'].rank(ascending=False)
    kpis_por_zona['ranking_eficiencia'] = kpis_por_zona['velocidad_promedio_mph'].rank(ascending=False)

    logging.info("Top 10 zonas más rentables...")
    top_rentables = kpis_por_zona.sort_values(by='ingreso_promedio_milla', ascending=False).head(10)
    top_rentables[['Zone', 'Borough', 'ingreso_promedio_milla', 'cantidad_viajes']]
except Exception as e:
    logging.error("Ocurrió un error durante la ejecución de la consulta", exc_info=True)
    

kpis_por_zona
```
